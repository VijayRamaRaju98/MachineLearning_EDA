```python
# we created function to return if the salary <=50k with 0, if salary >=50k with 1
def salary_data(sal):
    if sal == '<=50K':
        return 0
    else:
        return 1
```

```python
# applying salary_data function with salary income
data['encoded_income']=data['income'].apply(salary_data)
```

```python
# another way to repalce income values
data['income'] = data['income'].replace(['<=50K','>50K'],[0,1])
```
